# Employee Category Based on employee no.

```python
def Employee_category(value):
    if value in range(1,10):
        return '1'
    if value in range(11,50):
        return '2'
    elif value in range(51,100):
        return '3'
    elif value in range(101,250):
        return '4'
    elif value in range(251,500):
        return '5'
    elif value in range(510,1000):
        return '6'
    elif value in range(1001,5000):
        return '7'
    elif value in range(5001,10000):
        return '8'
    elif value > 10000:
        return '9'
    
df['Employee_category'] = df['total_employees'].map(Employee_category)
df
```


## No. of companies for each category
